```python
import pandas as pd
import numpy as np
from sklearn.metrics import confusion_matrix
from sklearn.metrics import precision_score
from sklearn.metrics import f1_score, recall_score, precision_score
from sklearn import metrics
```

```python
df_check = pd.read_csv(r'H:\IFT6759\20220410\ERM_Seed2_Densenet_Noaug\iwildcam_split_test_seed_2_epoch_best_pred.csv')
```

```python
a = pd.DataFrame(df_check.groupby('y_true'))
```

```python
class_label,num_img,correct_img,acc_img,f1_score_img=[],[],[],[],[]
precision_img,recall_img=[],[]
prediction=[]

df_class = pd.DataFrame()

for i in range(0,df_check.groupby('y_true').count().shape[0],1):
    
    df = a[1][i]
    y_true = df.y_true
    y_pred = df.y_pred
    ground_truth = a[1][i]['y_true'].unique()
    pred = a[1][i]['y_pred'].unique()
    prediction.append(pred.tolist()) 
    
    cm = confusion_matrix(y_true, y_pred)
    
    '''
    fp = cm.sum(axis=0) - np.diag(cm)  
    fn = cm.sum(axis=1) - np.diag(cm)
    tp = np.diag(cm)
    tn = cm.sum() - (fp + fn + tp)
    '''
    
    class_label.append(ground_truth[0].tolist())
    num_img.append(df.shape[0])
    correct_img.append(np.diag(cm).sum())
    
    precision_img.append(np.around(precision_score(y_true, y_pred, average="macro"),decimals=3)) 
    recall_img.append(np.around(recall_score(y_true, y_pred, average="macro"),decimals=3))
    acc_img.append(np.around(np.diag(cm).sum() / cm.sum(),decimals=3))
    f1_score_img.append(np.around(f1_score(y_true, y_pred, average='macro'),decimals=3))
    
    print("Results for Class {}".format(ground_truth[0]))
    print(metrics.classification_report(y_true, y_pred, digits=3))
    
df_class['Img_class'] = pd.Series(class_label)
df_class['Img_count'] = num_img 
df_class['True_Positive'] = correct_img 

'''
df_class['True_Positive'] = tp_img
df_class['True_Negative'] = tn_img
df_class['False_Positive'] = fp_img
df_class['False_Negative'] = fn_img
'''
df_class['Img_Precision'] = precision_img
df_class['Img_Recall'] = recall_img
df_class['Img_Acc'] = acc_img
df_class['Img_F1_macro'] = f1_score_img

df_class.to_csv(r'H:\IFT6759\20220410\ERM_Seed2_Densenet_Noaug\Summary_FracDataSet_DN_NoAug_ERM_OOD_Test.csv')
```
